```python
import pandas as pd
import numpy as np

# Create restaurant menu with 16 items across categories
menu_data = {
    'Item': [
        'Margherita Pizza', 'Pepperoni Pizza', 'Chicken Biryani', 'Mutton Biryani', 
        'Butter Chicken', 'Dal Makhani', 'Paneer Tikka', 'Fish Fry', 
        'Veg Manchurian', 'Chicken Fried Rice', 'Ice Cream', 'Gulab Jamun',
        'Chocolate Cake', 'Cold Coffee', 'Masala Dosa', 'Idli Sambhar'
    ],
    'Category': [
        'Pizza', 'Pizza', 'Biryani', 'Biryani', 
        'Curry', 'Curry', 'Starter', 'Starter', 
        'Chinese', 'Chinese', 'Dessert', 'Dessert',
        'Dessert', 'Beverage', 'South Indian', 'South Indian'
    ],
    'Price': [1299, 1499, 1199, 1599, 1399, 999, 1099, 1249, 899, 1049, 499, 599, 699, 449, 799, 649]
}

menu_df = pd.DataFrame(menu_data)
menu_df.index = range(1, len(menu_df) + 1)  # 1-based indexing for easy ordering
menu_df
```

```python
#This creates a clean Pandas DataFrame you can inspect with menu_df.head() or menu_df.describe().
```

```python
#Display Menu Function
```

```python
def display_menu():
    """Display the full menu in a formatted way"""
    print("=" * 50)
    print("üçΩÔ∏è  WELCOME TO BENGALURU RESTO - MENU üçΩÔ∏è")
    print("=" * 50)
    print(menu_df.to_string(index=True, index_names=False))
    print("\n" + "=" * 50)

display_menu()
```

```python
#Interactive Ordering System
```

```python
def take_order():
    """Interactive ordering with total calculation"""
    orders = []
    total = 0
    
    while True:
        display_menu()
        choice = input("\nEnter item number to order (or 'done' to finish): ").strip()
        
        if choice.lower() == 'done':
            break
            
        if choice.isdigit() and 1 <= int(choice) <= len(menu_df):
            item_num = int(choice)
            item_name = menu_df.loc[item_num, 'Item']
            price = menu_df.loc[item_num, 'Price']
            qty = int(input(f"Enter quantity for {item_name}: "))
            
            subtotal = price * qty
            total += subtotal
            
            order = {
                'Item': item_name,
                'Category': menu_df.loc[item_num, 'Category'],
                'Price': price,
                'Quantity': qty,
                'Subtotal': subtotal
            }
            orders.append(order)
            
            print(f"\n‚úÖ Added {qty} x {item_name} - Subtotal: ‚Çπ{subtotal:.2f}")
        else:
            print("‚ùå Invalid choice! Try a number from 1-16.")
    
    return pd.DataFrame(orders), total

# Test the ordering system
customer_orders, bill_total = take_order()
print(f"\nüí∞ Final Bill: ‚Çπ{bill_total:.2f}")
```

```python
#Analyze and Save Orders
```

```python
if not customer_orders.empty:
    # Display order summary
    print("\nüìã YOUR ORDER SUMMARY:")
    print(customer_orders.to_string(index=False))
    
    # Quick analytics
    print(f"\nüìä ORDER STATS:")
    print(f"Items ordered: {len(customer_orders)}")
    print(f"Total items: {customer_orders['Quantity'].sum()}")
    print(f"Avg price per item: ‚Çπ{customer_orders['Price'].mean():.2f}")
    print(f"Most expensive: ‚Çπ{customer_orders['Price'].max():.2f}")
    
    # Save to CSV (your new skill from past chats!)
    customer_orders.to_csv('restaurant_order.csv', index=False)
    print("\nüíæ Order saved to 'restaurant_order.csv'")
    
    # Category breakdown
    category_totals = customer_orders.groupby('Category')['Subtotal'].sum()
    print("\nüç¥ Breakdown by Category:")
    print(category_totals)
else:
    print("No orders placed!")
```
